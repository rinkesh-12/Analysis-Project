```python
# Draw Correlation Matrix
# sns_plot = sns.clustermap(df.corr(), cmap="rocket_r")
# print(df.corr())

# plt.matshow(df.corr()) # Correlation Method
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True) # Correlation Method
plt.show()
```

```python
# How Many People Have Heart Disease, And How Many Don't Have Heart Disease In This Dataset?
df['target'].value_counts()
```

```python
target_countplot = sns.countplot(x=df['target'], palette='rocket')
for bars in target_countplot.containers:
  target_countplot.bar_label(bars)
plt.title("Target Plot")
plt.show()
```

```python
# Find Count of  Male & Female in this Dataset
df['sex'].value_counts()
```

```python
gender_countplot = sns.countplot(x=df['sex'], palette='rocket')
for bars in gender_countplot.containers:
  gender_countplot.bar_label(bars)
plt.title("Gender Count")
plt.xticks([0, 1], ['Female', 'Male'])
plt.show()
```

```python
# Find Gender Distribution According to The Target Variable

gender_distribution = sns.countplot(x='sex', hue='target', data=df, palette='rocket')
for bars in gender_distribution.containers:
    gender_distribution.bar_label(bars)
plt.title("Gender Distribution According To Target Variable")
plt.xticks([0, 1], ['Female', 'Male'])
plt.legend(labels=['No-Disease', 'Disease'])
plt.show()
```

```python
# Check Age Distribution In The Dataset
age = sns.distplot(df['age'], bins=20, color='red')
# age = sns.histplot(df['age'], kde=True, bins=20, color='red')
plt.title("Age Distribution")
plt.show()
```

```python
# Check Chest Pain Type
    # 0: Typical Angina
    # 1: atypical Angina
    # 2: non-anginal pain
    # 3: asymptomatic

cp = sns.countplot(x=df['cp'], palette="rocket")
for bars in cp.containers:
    cp.bar_label(bars)
plt.title("Chest Pain Type")
plt.xticks([0, 1, 2, 3], ['Typical Angina', 'Atypical Angina', 'Non-anginal Pain', 'Asymptomatic'])
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.show()
```

```python
# Show The Chest Pain Distribution As Per Target Variable

age = sns.countplot(x=df['cp'], hue=df['target'], color='red')
plt.title("Chest Pain Distribution")
plt.xticks([0, 1, 2, 3], ['Typical Angina', 'Atypical Angina', 'Non-anginal Pain', 'Asymptomatic'])
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.legend(labels=['No-Disease', 'Disease'])
plt.show()
```

```python
 # Show Fasting Blood Sugar Distribution According To Target Variable

age = sns.countplot(x=df['fbs'], hue=df['target'], color='red')
plt.title("Fasting Blood Sugar Distribution")
# plt.xticks([0, 1], ['Typical Angina', 'Atypical Angina'])
plt.xlabel("Fasting Blood Sugar Type")
plt.ylabel("Count")
plt.legend(labels=['No-Disease', 'Disease'])
plt.show()
```

```python
# Check Resting Blood Pressure Distribution

df['trestbps'].hist()
plt.title("Resting Blood Pressure Distribution")
plt.show()
```

```python
# Compare Resting Blood Pressure As Per Sex Column
g = sns.FacetGrid(df, hue='sex', aspect=4)
g.map(sns.kdeplot, 'trestbps', shade=True)
plt.title("Resting Blood Pressure As Per Sex")
plt.legend(labels=['Female', 'Male'])
plt.show()

# sns.barplot(x=df['sex'], y=df['trestbps'], data=df, palette='rocket')
# plt.title("Resting Blood Pressure As Per Sex")
# plt.show()
```

```python
# Plot Continuous Variables
cate_val = []
cont_val = []

for column in df.columns:
  if df[column].nunique() <= 10:
      cate_val.append(column)
  else:
      cont_val.append(column)

print("Categorical Values:\n", cate_val)
print("\nContinuous Values:\n", cont_val)
```

```python
df.hist(cont_val, figsize=(15,6))
plt.title("Continuous Values Distribution")
plt.tight_layout()
plt.show()
```
